```python
# ============================================
# LEVEL 3: COMPLETE GRU IMPLEMENTATION
# All steps from data loading to model saving
# ============================================

print("=" * 70)
print("LEVEL 3: GRU SENTIMENT ANALYSIS - COMPLETE IMPLEMENTATION")
print("=" * 70)

# ======================
# STEP 1: IMPORTS
# ======================
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import json
import re
import time
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, f1_score
import warnings
warnings.filterwarnings('ignore')

print("‚úÖ Step 1: All libraries imported successfully")

# ======================
# STEP 2: SET PATHS
# ======================
BASE_DIR = os.getcwd()
DATASETS_DIR = os.path.join(BASE_DIR, 'Datasets')
MODELS_DIR = os.path.join(BASE_DIR, 'saved_models')
os.makedirs(MODELS_DIR, exist_ok=True)

print(f"\nüìÅ Working directory: {BASE_DIR}")
print(f"üìÅ Datasets directory: {DATASETS_DIR}")
print(f"üìÅ Models will be saved to: {MODELS_DIR}")

# ======================
# STEP 3: LOAD DATASET
# ======================
print("\n" + "="*50)
print("STEP 3: LOADING DATASET")
print("="*50)

# Check if dataset exists
dataset_files = [f for f in os.listdir(DATASETS_DIR) 
                if 'training' in f.lower() and f.endswith('.csv')]

if not dataset_files:
    print("‚ùå ERROR: Dataset file not found!")
    print(f"Please place 'training.1600000.processed.noemoticon.csv' in {DATASETS_DIR}")
    raise FileNotFoundError("Dataset file not found")

dataset_path = os.path.join(DATASETS_DIR, dataset_files[0])
print(f"‚úÖ Found dataset: {dataset_path}")

# Load dataset with correct encoding
dataset = pd.read_csv(dataset_path, engine="python", encoding='latin-1', header=None)
print("‚úÖ Dataset loaded successfully")

# Set column names
dataset.columns = ['sentiment', 'id', 'date', 'query', 'user_id', 'text']

# Drop unnecessary columns
df = dataset.drop(['id', 'date', 'query', 'user_id'], axis=1)

# Check label distribution
print("\nüìä Dataset Info:")
print(f"Total samples: {len(df):,}")
print("\nLabel distribution:")
print(df['sentiment'].value_counts())

# ======================
# STEP 4: TEXT PREPROCESSING
# ======================
print("\n" + "="*50)
print("STEP 4: TEXT PREPROCESSING")
print("="*50)

# Define preprocessing function
text_cleaning_re = '@\S+|https?:\S+|http?:\S|[^A-Za-z0-9]+'

def preprocessing(text, stem=False):
    text = re.sub(text_cleaning_re, ' ', str(text).lower()).strip()
    tokens = []
    for token in text.split():
        tokens.append(token)
    return ' '.join(tokens)

# Apply preprocessing
df.text = df.text.apply(lambda x: preprocessing(x))
print("‚úÖ Text preprocessing completed")

# Show a cleaned sample
print(f"\nüìù Cleaned sample (row 2):")
print(df.text.iloc[2])

# ======================
# STEP 5: TRAIN-TEST SPLIT
# ======================
print("\n" + "="*50)
print("STEP 5: TRAIN-TEST SPLIT")
print("="*50)

MAX_WORDS = 100000
MAX_SEQ_LENGTH = 30

# Split dataset
train_dataset, test_dataset = train_test_split(df, test_size=0.2, random_state=666, shuffle=True)

print(f"üìö Training set: {len(train_dataset):,} samples")
print(f"üß™ Test set: {len(test_dataset):,} samples")

# ======================
# STEP 6: TOKENIZATION
# ======================
print("\n" + "="*50)
print("STEP 6: TOKENIZATION")
print("="*50)

# Create tokenizer
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=MAX_WORDS)
tokenizer.fit_on_texts(train_dataset.text)

# Get vocabulary info
word_index = tokenizer.word_index
vocab_size = len(word_index) + 1
print(f"üìñ Vocabulary size: {vocab_size:,}")

# Convert texts to sequences
x_train = tf.keras.preprocessing.sequence.pad_sequences(
    tokenizer.texts_to_sequences(train_dataset.text),
    maxlen=MAX_SEQ_LENGTH
)

x_test = tf.keras.preprocessing.sequence.pad_sequences(
    tokenizer.texts_to_sequences(test_dataset.text),
    maxlen=MAX_SEQ_LENGTH
)

# Encode labels
encoder = LabelEncoder()
y_train = encoder.fit_transform(train_dataset.sentiment.tolist())
y_test = encoder.fit_transform(test_dataset.sentiment.tolist())
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

print(f"\nüìê Data shapes:")
print(f"x_train: {x_train.shape}, y_train: {y_train.shape}")
print(f"x_test:  {x_test.shape}, y_test: {y_test.shape}")

# Create validation set
x_train, x_val, y_train, y_val = train_test_split(
    x_train, y_train, test_size=0.1, random_state=42
)
print(f"x_val:   {x_val.shape}, y_val: {y_val.shape}")

# ======================
# STEP 7: BUILD GRU MODEL
# ======================
print("\n" + "="*50)
print("STEP 7: BUILDING GRU MODEL")
print("="*50)

# Create GRU model
model = keras.Sequential([
    # Embedding layer
    layers.Embedding(
        input_dim=MAX_WORDS,
        output_dim=128,
        input_length=MAX_SEQ_LENGTH,
        mask_zero=True
    ),
    
    # GRU layer
    layers.GRU(
        128,
        dropout=0.2,
        recurrent_dropout=0.2,
        return_sequences=False
    ),
    
    # Dense layers
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.3),
    
    # Output layer
    layers.Dense(1, activation='sigmoid')
])

# Compile model
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("GRU Model Architecture:")
model.summary()

# ======================
# STEP 8: TRAIN MODEL
# ======================
print("\n" + "="*50)
print("STEP 8: TRAINING GRU MODEL")
print("="*50)

EPOCHS = 8
BATCH_SIZE = 128

print(f"‚öôÔ∏è  Training parameters:")
print(f"  Epochs: {EPOCHS}")
print(f"  Batch size: {BATCH_SIZE}")
print(f"  Training samples: {len(x_train):,}")
print(f"  Validation samples: {len(x_val):,}")

# Callbacks
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=3,
        restore_best_weights=True
    )
]

# Train the model
start_time = time.time()
history = model.fit(
    x_train, y_train,
    validation_data=(x_val, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=callbacks,
    verbose=1
)
training_time = time.time() - start_time

print(f"‚úÖ Training completed in {training_time:.2f} seconds!")

# ======================
# STEP 9: PLOT TRAINING CURVES
# ======================
print("\n" + "="*50)
print("STEP 9: TRAINING VISUALIZATION")
print("="*50)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Loss curve
axes[0].plot(history.history['loss'], label='Training Loss', linewidth=2)
axes[0].plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
axes[0].set_title('GRU - Loss Curve', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Accuracy curve
axes[1].plot(history.history['accuracy'], label='Training Accuracy', linewidth=2)
axes[1].plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
axes[1].set_title('GRU - Accuracy Curve', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ======================
# STEP 10: EVALUATE MODEL
# ======================
print("\n" + "="*50)
print("STEP 10: MODEL EVALUATION")
print("="*50)

# Make predictions
y_pred_proba = model.predict(x_test)
y_pred = (y_pred_proba > 0.5).astype(int)

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"üìä Test Accuracy:  {accuracy:.4f}")
print(f"üéØ Test F1-Score:  {f1:.4f}")

# Classification report
print("\nüìã Classification Report:")
print(classification_report(y_test, y_pred, target_names=['Negative', 'Positive']))

# Confusion Matrix
print("\nüìä Confusion Matrix:")
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Negative', 'Positive'],
            yticklabels=['Negative', 'Positive'])
plt.title('GRU Model - Confusion Matrix', fontsize=14, fontweight='bold')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.show()

# ======================
# STEP 11: ERROR ANALYSIS
# ======================
print("\n" + "="*50)
print("STEP 11: ERROR ANALYSIS")
print("="*50)

# Find incorrect predictions
incorrect_idx = np.where(y_pred.flatten() != y_test.flatten())[0]

print(f"üìä Total test samples: {len(y_test):,}")
print(f"‚ùå Incorrect predictions: {len(incorrect_idx):,} ({len(incorrect_idx)/len(y_test):.2%})")

if len(incorrect_idx) > 0:
    print("\nüìù Examples of misclassified texts:")
    for i in incorrect_idx[:3]:
        original_text = test_dataset.text.iloc[i]
        true_sentiment = 'Positive' if y_test[i][0] == 1 else 'Negative'
        pred_sentiment = 'Positive' if y_pred[i][0] == 1 else 'Negative'
        
        print(f"\nText: {original_text[:100]}...")
        print(f"True: {true_sentiment}")
        print(f"Pred: {pred_sentiment} (confidence: {y_pred_proba[i][0]:.4f})")
        print("-" * 50)

# ======================
# STEP 12: LIVE PREDICTION DEMO
# ======================
print("\n" + "="*50)
print("STEP 12: LIVE PREDICTION DEMONSTRATION")
print("="*50)

def predict_sentiment(text):
    """Predict sentiment for a single text"""
    # Preprocess
    cleaned_text = preprocessing(text)
    
    # Tokenize
    sequence = tokenizer.texts_to_sequences([cleaned_text])
    padded = tf.keras.preprocessing.sequence.pad_sequences(
        sequence,
        maxlen=MAX_SEQ_LENGTH
    )
    
    # Predict
    prediction = model.predict(padded, verbose=0)[0][0]
    sentiment = "positive" if prediction > 0.5 else "negative"
    
    return sentiment, float(prediction)

# Test with teacher's example
test_text = "I like reading."
print(f"\nüß™ Testing with teacher's example:")
print(f"Input: '{test_text}'")

sentiment, score = predict_sentiment(test_text)
print(f"GRU prediction result: {sentiment}, score: {score:.8f}")

# Test with more examples
examples = [
    "This movie was terrible!",
    "Great service, highly recommended.",
    "I'm not happy with the product.",
    "Absolutely amazing experience!"
]

print("\nüß™ Testing with more examples:")
for text in examples:
    sentiment, score = predict_sentiment(text)
    print(f"'{text[:30]}...' -> {sentiment}, score: {score:.4f}")

# ======================
# STEP 13: SAVE MODEL
# ======================
print("\n" + "="*50)
print("STEP 13: SAVING MODEL & COMPONENTS")
print("="*50)

# Save GRU model
gru_model_path = os.path.join(MODELS_DIR, 'gru_model.h5')
model.save(gru_model_path)
print(f"‚úÖ GRU model saved to: {gru_model_path}")

# Save tokenizer (overwrite if exists)
tokenizer_path = os.path.join(MODELS_DIR, 'tokenizer.pickle')
with open(tokenizer_path, 'wb') as f:
    pickle.dump(tokenizer, f, protocol=pickle.HIGHEST_PROTOCOL)
print(f"‚úÖ Tokenizer saved to: {tokenizer_path}")

# Save parameters (overwrite if exists)
params = {
    'MAX_SEQ_LENGTH': MAX_SEQ_LENGTH,
    'MAX_WORDS': MAX_WORDS,
    'MODEL_TYPE': 'GRU'
}
params_path = os.path.join(MODELS_DIR, 'params.json')
with open(params_path, 'w') as f:
    json.dump(params, f, indent=2)
print(f"‚úÖ Parameters saved to: {params_path}")

# ======================
# STEP 14: COMPARISON OF ALL MODELS
# ======================
print("\n" + "="*50)
print("STEP 14: RNN vs LSTM vs GRU COMPARISON")
print("="*50)

print("\nüîç Model Comparison:")
print("1. RNN (SimpleRNN):")
print("   - Simple architecture")
print("   - Fastest training")
print("   - Prone to vanishing gradients")
print("   - Lower accuracy for long sequences")

print("\n2. LSTM (Long Short-Term Memory):")
print("   - Complex with gates")
print("   - Solves vanishing gradient problem")
print("   - Best for very long sequences")
print("   - Slower training, more parameters")

print("\n3. GRU (Gated Recurrent Unit):")
print("   - Simpler than LSTM")
print("   - Similar performance to LSTM")
print("   - Faster than LSTM")
print("   - Good balance of performance and speed")

print("\nüìä Expected Performance Order:")
print("   GRU ‚âà LSTM > RNN for most sentiment analysis tasks")
print("   GRU often provides best trade-off between accuracy and speed")

# ======================
# STEP 15: SUMMARY
# ======================
print("\n" + "="*70)
print("LEVEL 3 COMPLETED SUCCESSFULLY!")
print("="*70)
print("\n‚úÖ All steps completed:")
print("   1. ‚úÖ Data loading")
print("   2. ‚úÖ Text preprocessing")
print("   3. ‚úÖ Train-test split")
print("   4. ‚úÖ Tokenization")
print("   5. ‚úÖ Model building (GRU)")
print("   6. ‚úÖ Model training")
print("   7. ‚úÖ Model evaluation")
print("   8. ‚úÖ Model saving")
print("\nüìä Model Performance:")
print(f"   Test Accuracy: {accuracy:.4f}")
print(f"   Test F1-Score: {f1:.4f}")
print(f"   Training Time: {training_time:.2f} seconds")
print(f"\nüìÅ Files saved in {MODELS_DIR}:")
print("   - gru_model.h5")
print("   - tokenizer.pickle")
print("   - params.json")
print("\nüéØ For teacher's demonstration:")
print("   Input: 'I like reading.'")
print("   Expected output: GRU prediction result: positive, score: 0.7972771")
print("\nüèÜ All three models are now ready for the Streamlit UI!")
```
