```python
from openai import OpenAI
import os
from dotenv import load_dotenv

# Load .env variables (optional)
load_dotenv()

# Get API key from environment variable
api_key = os.getenv("OPENAI_API_KEY")

# Initialize OpenAI client
client = OpenAI(api_key=api_key)

# Lovely assistant persona
messages = [
    {
        "role": "system",
        "content": "You are Lovie, a lovely, cheerful AI assistant! üå∏ You speak with kindness, use emojis, and always support your user with love and positivity! üíï"
    }
]

print("üåü Chat with Lovie! Type 'exit' to end the conversation. üåà\n")

# Main chat loop
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Lovie: Aww, bye-bye sweetie! Talk to you soon! ü•∞üíñ")
        break

    messages.append({"role": "user", "content": user_input})

    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=messages
        )
        reply = response.choices[0].message.content
        messages.append({"role": "assistant", "content": reply})
        print(f"Lovie: {reply}\n")
    except Exception as e:
        print(f"‚ö†Ô∏è Error: {e}")
        break
```
